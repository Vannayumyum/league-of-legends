# Web scrapping on op.gg

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import urllib.request
```

```python
url='https://www.op.gg/leaderboards/tier?region=euw'
```

```python
page=urllib.request.urlopen(url,timeout=5)
```

```python
soup=bs(page)
```

When you are on the web page, right click and inspect it. You will get access to the html  
Right click on the element you need (for ex: the first challengers of euw) and right click again and inspect it.  
You can do the same for any features you want on the web page.

```python
#Extracting summoner's names
name= soup.find_all('strong',{'class': 'summoner-name'})
```

```python
sum_name=[]
for e in name :
    e=e.text
    e=e.replace('\n','')
    sum_name.append(e)
```

```python
#Extracting number of LP
lp=soup.find_all('td',{'class': 'css-1oq0a1y e1g3wlsd6'})
```

```python
lp_list=[]
for e in lp :
    e=e.text
    e=e.replace('\n','')
    lp_list.append(e)
```

```python
#Extracting win/rate
ratio=soup.find_all('div',{'class': 'text'})
wr=[]
for e in ratio :
    e=e.text
    e=e.replace('\n','')
    wr.append(e)
```

```python
#We create a dataframe
chall_ladder=[]
for i in range(len(sum_name)):
    info_row={}
    info_row['Summoner name']=sum_name[i]
    info_row['LP']=lp_list[i]
    info_row['Winrate']=wr[i]
    chall_ladder.append(info_row)
```

```python

df=pd.DataFrame(chall_ladder)
```
